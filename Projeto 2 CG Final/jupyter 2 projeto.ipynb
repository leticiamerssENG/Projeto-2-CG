```python
import matplotlib.pyplot as plt
import numpy as np
from PIL import Image

img1 = np.array(Image.open('img_ruido.png'))

m,n = img1.shape

img_new = np.zeros([m,n])

for i in range (1, m-1):
    for j in range (1, n-1):
        temp = [img1[i-1, j-1], 
                img1[i-1, j], 
                img1[i-1, j+1], 
                img1[i, j-1], 
                img1[i, j], 
                img1[i, j+1], 
                img1[i+1, j-1], 
                img1[i+1, j], 
                img1[i+1, j+1]]
        temp = sorted(temp)
        img_new[i, j] = temp [4]
                                                                                                                             
img_new = img_new.astype(np.uint8)

plt.figure(figsize=(16, 16))
plt.subplot(2, 2, 1)
plt.imshow(img1, cmap = 'gray')
plt.subplot(2, 2, 2)
plt.imshow(img_new, cmap = 'gray')
```
